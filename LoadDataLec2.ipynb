```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

DATADIR="/content/drive/MyDrive/Machine_Learning_Datas/PetImages"
CATEGORIES=["Dog","Cat"]

for category in CATEGORIES:
  path=os.path.join(DATADIR,category)  #this is the path to cats and dogs dir
  for img in os.listdir(path):
    img_array=cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_array, cmap="gray")
    plt.show()
    break
  break
```

```python
IMG_SIZE=50

new_array=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap="gray")
plt.show
```

```python
training_data=[]

def create_training_data():
    for category in CATEGORIES:
      path=os.path.join(DATADIR,category)  #this is the path to cats and dogs dir
      class_num= CATEGORIES.index(category)
      for img in os.listdir(path):
        try:
          img_array=cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
          new_array=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
          training_data.append([new_array, class_num])
        except Exception as e:
          pass
create_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
for sample in training_data:
  print(sample[1])
```

```python
X=[]
y=[]
```

```python
for features,label in training_data:
  X=np.append(X,features)
  y=np.append(y,label) #this is how to append it as if we did y.append(label) it would have caused
  #an error by showing that numpy.ndarray object has no attribute append
  X=np.array(X).reshape(-1,IMG_SIZE, IMG_SIZE, 1) # 1 is for grayscale
```

```python
import pickle

pickle_out=open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out=open("y.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
