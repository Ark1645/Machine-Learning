```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
# df=pd.read_csv("/content/drive/MyDrive/canada_per_capita_income.csv")
```

```python
df=pd.read_excel("/content/drive/MyDrive/canada_per_capita_incomeA.xlsx")
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[{'year'}],df.per_capita_income_in_dollars)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per capita income in US($)")
plt.scatter(df.year,df.per_capita_income_in_dollars, color='red', marker="+")
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
